# Attempting to load dataframe with data dictionary declared

```python
clms_list = list(all_data)
dtype_list = [all_data[i].dtype.name for i in all_data]

data_dict = dict(zip(clms_list, dtype_list))
```

```python
project = redcap.Project(config['api_url'], config["api_token"])
```

```python
%%time
all_data = project.export_records(format="df", df_kwargs = {"dtype":data_dict})
all_data.tail(2)
```
